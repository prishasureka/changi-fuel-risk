### ***Data Collection***

**Fetch World Bank GDP data via Python**

```python
import requests
import pandas as pd

indicator = "NY.GDP.PCAP.CD"
url = f"https://api.worldbank.org/v2/country/all/indicator/{indicator}"

params = {
    "format": "json",
    "date": "2005:2024",
    "per_page": 20000  # try to get everything in one go
}

response = requests.get(url, params=params)
data = response.json()

# World Bank returns: [metadata, actual data]
rows = data[1]

# Flatten into a clean dataframe
gdp_raw = pd.json_normalize(rows)

# Keep only useful columns (optional but recommended)
gdp_raw = gdp_raw[[
    "country.value",
    "countryiso3code",
    "date",
    "value"
]]

# Rename columns to match your wbdata version
gdp_raw.columns = ["country", "iso3", "year", "gdp_per_capita"]

# Convert year properly
gdp_raw["year"] = gdp_raw["year"].astype(int)

print(gdp_raw.head(10))
print(f"Shape: {gdp_raw.shape}")

# Save immediately
gdp_raw.to_csv("../data/raw/world_bank_gdp_raw.csv", index=False)
print("Saved.")
```
